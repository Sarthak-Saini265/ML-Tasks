## VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data.drop('Previous Scores', axis=1)[['Hours Studied', 'Sleep Hours', 'Sample Question Papers Practiced' ]]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

## Creating Dummies

```python
data_with_dummies = pd.get_dummies(data, drop_first=True)
```

```python
data_with_dummies['Extracurricular Activities_Yes'] = data_with_dummies['Extracurricular Activities_Yes'].astype(int)
```

```python
data_with_dummies
```

## Standardization

```python
targets = data_with_dummies['Performance Index']
inputs = data_with_dummies.drop(['Performance Index'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state = 44)
```

## Regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
y_hat
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
plt.show()
```

```python
reg.score(x_test, y_test)
```

```python
sns.distplot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.show()
```

```python
df_predict = pd.DataFrame(y_hat_test, columns = ['Predictions'])
df_predict.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_predict['Target'] = y_test
df_predict
```

```python
df_predict['Residual'] = df_predict['Target'] - df_predict['Predictions']
```

```python
df_predict
```

```python
df_predict['Difference%'] = np.absolute(df_predict['Residual']/df_predict['Target']*100)
```
