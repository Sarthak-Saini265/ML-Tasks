```python
mean_setosa = [sum(cluster_setosa.iloc[:, i])/len(cluster_setosa) for i in range(4)]
mean_versicolour = [sum(cluster_versicolour.iloc[:, i])/len(cluster_versicolour) for i in range(4)]
mean_virginica = [sum(cluster_virginica.iloc[:, i])/len(cluster_virginica) for i in range(4)]
mean_setosa
```

```python
mean_versicolour
```

```python
mean_virginica
```

```python
def distance(row):
    distance = {}
    distance['Setosa'] = (math.sqrt((row[0] - mean_setosa[0])**2 + (row[1] - mean_setosa[1])**2 + (row[2] - mean_setosa[2])**2 + (row[3] - mean_setosa[3])**2))
    distance['Versicolour'] = (math.sqrt((row[0] - mean_versicolour[0])**2 + (row[1] - mean_versicolour[1])**2 + (row[2] - mean_versicolour[2])**2 + (row[3] - mean_versicolour[3])**2))
    distance['Virginica'] = (math.sqrt((row[0] - mean_virginica[0])**2 + (row[1] - mean_virginica[1])**2 + (row[2] - mean_virginica[2])**2 + (row[3] - mean_virginica[3])**2))
    return list(distance.keys())[list(distance.values()).index(min(distance.values()))]
```
