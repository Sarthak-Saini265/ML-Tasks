```python
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
file = pd.read_csv('Iris-Dataset.csv')
data = file.copy()
file
```

```python
cluster_setosa = data[data['y'] == 'Iris-Setosa']
cluster_versicolour = data[data['y'] == 'Iris-Versicolour']
cluster_virginica = data[data['y'] == 'Iris-Virginica']
cluster_setosa
```

```python
plt.scatter(cluster_setosa['x1'], cluster_setosa['x2'])
plt.scatter(cluster_versicolour['x1'], cluster_versicolour['x2'], color='orange')
plt.scatter(cluster_virginica['x1'], cluster_virginica['x2'],color='limegreen')
plt.show()
```

```python
mean_setosa = [sum(cluster_setosa.iloc[:, i])/len(cluster_setosa) for i in range(4)]
```

```python
mean_setosa
```

```python
sum(data.iloc[:, 0])
```

```python
mean_versicolour = [sum(cluster_versicolour.iloc[:, i])/len(cluster_versicolour) for i in range(4)]
mean_versicolour
```

```python
mean_virginica = [sum(cluster_virginica.iloc[:, i])/len(cluster_virginica) for i in range(4)]
mean_virginica
```

```python
random_point = [7.3, 2.5, 5.7, 3.9]
random_point
```

```python
distance = {}
distance['Setosa'] = (math.sqrt((random_point[0] - mean_setosa[0])**2 + (random_point[1] - mean_setosa[1])**2 + (random_point[2] - mean_setosa[2])**2 + (random_point[3] - mean_setosa[3])**2))
distance['Versicolour'] = (math.sqrt((random_point[0] - mean_versicolour[0])**2 + (random_point[1] - mean_versicolour[1])**2 + (random_point[2] - mean_versicolour[2])**2 + (random_point[3] - mean_versicolour[3])**2))
distance['Virginica'] = (math.sqrt((random_point[0] - mean_virginica[0])**2 + (random_point[1] - mean_virginica[1])**2 + (random_point[2] - mean_virginica[2])**2 + (random_point[3] - mean_virginica[3])**2))
```

```python
distance
```

```python
min(distance.values())
```

```python
min_distance = list(distance.keys())[list(distance.values()).index(min(distance.values()))]
```

```python
print(f'The point lies in the {min_distance} cluster')
```
