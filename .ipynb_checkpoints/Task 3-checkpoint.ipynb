```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('Iris-Dataset.csv')
data = file.copy()
data
```

```python
indices_to_select = np.random.choice(data[data['y'] == 'Iris-Setosa'].index, size=40, replace=False)
cluster_setosa = data.loc[indices_to_select]
data = data.drop(indices_to_select)
```

```python
cluster_setosa
```

```python
indices_to_select = np.random.choice(data[data['y'] == 'Iris-Versicolour'].index, size=40, replace=False)
cluster_versicolour = data.loc[indices_to_select]
data = data.drop(indices_to_select)
```

```python
cluster_versicolour
```

```python
indices_to_select = np.random.choice(data[data['y'] == 'Iris-Virginica'].index, size=40, replace=False)
cluster_virginica = data.loc[indices_to_select]
data = data.drop(indices_to_select)
```

```python
cluster_virginica
```

```python
plt.scatter(cluster_setosa['x1'], cluster_setosa['x2'])
plt.scatter(cluster_versicolour['x1'], cluster_versicolour['x2'], color='orange')
plt.scatter(cluster_virginica['x1'], cluster_virginica['x2'],color='limegreen')
plt.show()
```

```python
mean_setosa = [sum(cluster_setosa.iloc[:, i])/len(cluster_setosa) for i in range(4)]
mean_versicolour = [sum(cluster_versicolour.iloc[:, i])/len(cluster_versicolour) for i in range(4)]
mean_virginica = [sum(cluster_virginica.iloc[:, i])/len(cluster_virginica) for i in range(4)]
mean_setosa
```
